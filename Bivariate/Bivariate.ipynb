```python
def quanqual(dataset):
    quan=[]
    qual=[]
    for ColumnName in dataset.columns:
           # print(ColumnName)
            if (dataset[ColumnName].dtype=='O'):
#print("it is quantitative variable / column")   #numbers may be discrete or continous
                qual.append(ColumnName)
            else:
               # print("it is qualitative variable / column")      # Categorical data - nominal, or ordinal
                quan.append(ColumnName)
    return quan,qual 
```

```python
quan,qual = quanqual(dataset)
```

```python
# Computes the Variance Inflation Factor (VIF) for each numeric column
#from statsmodels.stats.ouliers_influence import varience_inflation_factor

from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(x_input):
              
        vif_data = pd.DataFrame()
        vif_data["varibles"] = x_input.columns
        vif_data["VIF"] = [ 
            variance_inflation_factor(x_input.values, i ) 
            for i in range(x_input.shape[1])
        ] 
        return vif_data
```

```python
#Calculate two way analysis"
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example dataset
data = pd.DataFrame({
    'Marks': [70, 75, 80, 72, 78, 85, 90, 88, 92, 95, 99, 91],
    'Gender': ['M', 'M', 'M', 'M', 'F', 'F', 'F', 'F', 'M', 'M', 'F', 'F'],
    'Teaching_Method': ['A', 'A', 'B', 'B', 'A', 'A', 'B', 'B', 'A', 'B', 'A', 'B']
})

# Two-way ANOVA model
model = ols('Marks ~ C(Gender) + C(Teaching_Method) + C(Gender):C(Teaching_Method)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
#Term	p-value	Interpretation
#C(Gender)	0.20	Since p > 0.05 â†’ Gender has no significant effect on Marks
#C(Teaching_Method)	0.63	Teaching method also has no significant effect
#C(Gender):C(Teaching_Method)	0.93	No significant interaction effect between Gender and Method
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.factorplots import interaction_plot

fig = interaction_plot(data['Teaching_Method'], data['Gender'], data['Marks'],
                       colors=['red','blue'], markers=['D','^'], ms=10)
plt.title("Interaction Plot: Gender Ã— Teaching Method")
plt.xlabel("Teaching Method")
plt.ylabel("Mean Marks")
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example DataFrame
data = pd.DataFrame({
    'exercise_type': ['Aerobic', 'Aerobic', 'Strength', 'Strength', 'Aerobic', 'Strength'] * 10,
    'age_group': ['Teen', 'Adult', 'Senior', 'Teen', 'Adult', 'Senior'] * 10,
    'fitness_change': [4.2, 5.1, 3.5, 4.8, 5.0, 3.3] * 10
})

# Two-way ANOVA model
model = ols('fitness_change ~ C(exercise_type) + C(age_group) + C(exercise_type):C(age_group)', data=data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pointplot(x='age_group', y='fitness_change', hue='exercise_type', data=data, 
              capsize=0.1, errwidth=1.5, palette='Set2')
plt.title("Interaction between Exercise Type and Age Group on Fitness Change")
plt.show()
```
