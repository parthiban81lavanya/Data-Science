```python
#Call the function Quan qual and assign to a valraible.
def quanqual(dataset):
    quan=[]
    qual=[]
    for ColumnName in dataset.columns:
           # print(ColumnName)
            if (dataset[ColumnName].dtype=='O'):
#print("it is quantitative variable / column")   #numbers may be discrete or continous
                qual.append(ColumnName)
            else:
               # print("it is qualitative variable / column")      # Categorical data - nominal, or ordinal
                quan.append(ColumnName)
    return quan,qual
```

```python
quan,qual = quanqual(dataset) 
```

```python
dataset[quan] 
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
```

```python
imp
```

```python
df = imp.transform(dataset[quan])
```

```python
df
```

```python
#convert this to a table:
df_quan_table =  pd.DataFrame(df,columns=quan)
```

```python
df_quan_table
```

```python
df_qual_table =  dataset[qual] 
```

```python
frames = (df_quan_table,df_qual_table)  # join 2 data frames
pre_processed_data = pd.concat(frames,axis=1)  #Full table qual and quan udpated. Add axis for column wise concatenation!!
pre_processed_data
```
