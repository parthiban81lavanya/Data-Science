```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Filter students not placed
not_placed = dataset[dataset['status'] == 'Not Placed']

# 2. Identify possible reasons (threshold = 60%)
low_ssc = not_placed[not_placed['ssc_p'] < 60]
low_hsc = not_placed[not_placed['hsc_p'] < 60]
low_degree = not_placed[not_placed['degree_p'] < 60]
low_mba = not_placed[not_placed['mba_p'] < 60]
no_workexp = not_placed[not_placed['workex'] == 'No']
low_etest = not_placed[not_placed['etest_p'] < 60]

# 3. Create summary table
reason_summary = {
    'Low SSC % (<60)': len(low_ssc),
    'Low HSC % (<60)': len(low_hsc),
    'Low Degree % (<60)': len(low_degree),
    'Low MBA % (<60)': len(low_mba),
    'No Work Experience': len(no_workexp),
    'Low E-Test Score (<60)': len(low_etest)
}

reason_df = pd.DataFrame(list(reason_summary.items()), columns=['Reason', 'Count'])

# 4. Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(reason_df['Reason'], reason_df['Count'])
plt.title('Reasons for Students Not Being Placed', fontsize=14)
plt.xlabel('Reason', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

############################################################3333
total_not_placed = len(not_placed)
# 3. Create summary table (in percentage)
reason_summary = {
    'Low SSC % (<60)': (len(low_ssc) / total_not_placed) * 100,
    'Low HSC % (<60)': (len(low_hsc) / total_not_placed) * 100,
    'Low Degree % (<60)': (len(low_degree) / total_not_placed) * 100,
    'Low MBA % (<60)': (len(low_mba) / total_not_placed) * 100,
    'No Work Experience': (len(no_workex) / total_not_placed) * 100,
    'Low E-Test Score (<60)': (len(low_etest) / total_not_placed) * 100
}

reason_df = pd.DataFrame(list(reason_summary.items()), columns=['Reason', 'Percentage'])

# 4. Plot bar chart (percentage)
plt.figure(figsize=(10, 6))
bars = plt.bar(reason_df['Reason'], reason_df['Percentage'])

# Add percentage labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.title('Reasons for Students Not Being Placed (%)', fontsize=14)
plt.xlabel('Reason', fontsize=12)
plt.ylabel('Percentage of Not Placed Students', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Standard Normal Discribution:
def sdNDgraph(dataset):
    import seaborn as sn
    mean=dataset.mean()
    sd=dataset.std()

    values=[i for i in dataset]
    z_score= [((j-mean)/sd) for j in values] 

    sns.displot(z_score,kde=True)
   # sum(z_score/len(z_score))
    result = sum([x / len(z_score) for x in z_score])
    print(result)
```

```python
# 10)What is the probability Density Function of the salary range from 700000 to 900000?

def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob

```
