```python
# Dropping columns which are not required for further analysis
#dataset_raw1.drop(['URL', 'PAGE NO', 'CITY'], axis = 1, inplace=True)
dataset_raw_chn.drop(['Zomato URL', 'Dining Rating Count', 'Delivery Rating Count'], axis=1, errors='ignore', inplace=True)
```

```python
# Replace Null records with NA
dataset_raw_chn.fillna('NA', inplace=True)
dataset_raw_chn.replace("NA", np.nan, inplace=True)
dataset_raw_chn.dropna(inplace=True)
```

```python
useless_data = (dataset_raw_chn['Dining Rating'] == 0.0) | (dataset_raw_chn['Delivery Rating'] == 0)
dataset_raw_chn[useless_data]
```

```python
dataset_raw_chn['Dining Rating'].replace(to_replace=['-','None'], value='0', inplace=True)
dataset_raw_chn
```

```python
dataset_raw_chn = dataset_raw_chn.drop_duplicates()
```

```python
dataset_raw_chn = dataset_raw_chn[~useless_data]
```

```python
dataset_raw_chn['Cuisine'] = dataset_raw_chn['Cuisine'].astype(str).str.replace(r"[\[\]']", '', regex=True)
dataset_raw_chn['Top Dishes'] = dataset_raw_chn['Top Dishes'].astype(str).str.replace(r"[\[\]']", '', regex=True)
dataset_raw_chn['Features'] = dataset_raw_chn['Features'].astype(str).str.replace(r"[\[\]']", '', regex=True)
#df['column'] = df['column'].str.replace(r"[\[\]']", '', regex=True)
```

```python
zomato_df = dataset_raw_chn.copy()
```

```python
zomato_df
```

```python
import plotly.express as px
fig = px.histogram(zomato_df, x='Delivery Rating', color='Cuisine',
 title= 'No. of Restaurants by Cuisine',
 labels={'Location':'Location'})
fig.show()
```

```python
rating_type_df = zomato_df['Delivery Rating'].value_counts().reset_index()
rating_type_df.rename(columns={'Delivery Rating':'Delivery Rating', 'count':'COUNT OF RESTAURANTS'}, inplace=True)
rating_type_df
```

```python
#Q1) How many restaurants are in Chennai for each type of cuisine?
```

```python
#Q2 What are the percentage of restaurants by Dinning and Delivery Rating Type in Chennai?
fig = px.pie(rating_type_df, names='Delivery Rating', values='COUNT OF RESTAURANTS', color='Delivery Rating',
 title='Percentage of Restaurants by Rating Type').update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
rating_dine_df = zomato_df['Dining Rating'].value_counts().reset_index()
rating_dine_df.rename(columns={'Dining Rating':'Dining Rating', 'count':'COUNT OF RESTAURANTS'}, inplace=True)
rating_dine_df
```

```python
fig = px.pie(rating_dine_df, names='Dining Rating', values='COUNT OF RESTAURANTS', color='Dining Rating',
 title='Percentage of Restaurants by Rating').update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
loc_type_df = zomato_df['Location'].value_counts().reset_index()
loc_type_df.rename(columns={'Location':'Location', 'count':'COUNT OF RESTAURANTS'}, inplace=True)
loc_type_df
```

```python
fig = px.pie(loc_type_df, names='Location', values='COUNT OF RESTAURANTS', color='Location',
 title='Restaurants by Location').update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
#Q5) Which places have the highest rated restaurant for each Cuisine Type in Mumbai?
highest_rated_df = zomato_df[zomato_df['Delivery Rating'] >= 4.5]
highest_rated_df
```

```python
fig = px.histogram(
    highest_rated_df,
    x='Location',
    color='Cuisine',
    title='No. of Best Restaurant for each Cuisine Type by Places'
)
fig.update_xaxes(categoryorder="total descending")
fig.show()
```

```python
# Q6) What is the Avg Price Distibution of highest rated restaurant for each Cuisine Type in Chennai?
highest_rated_price_df = highest_rated_df.groupby(by=['Location', 'Cuisine'])['Price for 2'].mean().reset_index()
highest_rated_price_df.head()
```

```python
fig = px.scatter(highest_rated_price_df, x="Location", y="Price for 2", color="Cuisine", symbol="Cuisine",
 title=' Avg Price Distibution of High rated restaurant for each Cuisine Type').update_traces(marker_size=12)
fig.show()
```

```python
#Q7) Which areas have a large number of Chinese Restaurant Market?
icecream_df = zomato_df[zomato_df['Cuisine'].str.contains('Ice Cream')]
icecream_df
```

```python
chinese_rest_df = (
    icecream_df
    .groupby(by='Location')
    .agg({'Name of Restaurant': 'count', 'Price for 2': 'mean'})
    .rename(columns={'Name of Restaurant': 'COUNT OF RESTAURANTS', 'Price for 2': 'Average_Price for 2'})
)
chinese_rest_df = chinese_rest_df.sort_values('COUNT OF RESTAURANTS', ascending=False).head(25)
chinese_rest_df.head()
```

```python
fig = px.bar(
    chinese_rest_df,
    x=chinese_rest_df.index,
    y='COUNT OF RESTAURANTS',
    color='Average_Price for 2',
    title='Number of Chinese Restaurants by Region'
)
fig.show()
```

```python
# Q9 IS ther a relation between region and price

region_price_df = zomato_df.groupby(['Location'])['Price for 2'].mean().reset_index()
region_price_df
```

```python
fig = px.scatter(region_price_df, x="Location", y="Price for 2").update_traces(marker_size=8)
fig.update_xaxes(
    tick0=0,
    dtick=1000,
    showgrid=True,
    gridcolor='lightgray',   # grid line color
  )
fig.show()
```

```python
# Q10 IS ther a relation between Feature and price

fe_price_df = zomato_df.groupby(['Features'])['Price for 2'].mean().reset_index()
fe_price_df
```

```python
fig = px.scatter(fe_price_df, x="Features", y="Price for 2").update_traces(marker_size=8)
fig.update_xaxes(
    tick0=0,
    dtick=2000,
    showgrid=True,
    gridcolor='lightgray',   # grid line color
  )
fig.show()
```
